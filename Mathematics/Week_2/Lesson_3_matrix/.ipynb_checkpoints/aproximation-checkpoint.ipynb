```python
import matplotlib
import numpy as np
from scipy.linalg import solve
x = np.arange(1., 15, 0.1)
f = np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
x1 = np.array([1, 15])
f1 = np.sin(x1 / 5) * np.exp(x1 / 10) + 5 * np.exp(-x1 / 2)
matplotlib.pyplot.plot(x, f, x1, f1)
```

```python
f1
```

```python
A = np.array([[1, 1], [1, 15]])
x = solve(A, f1)
x
```

```python
x_test = np.array([1, 15])
f_1n = x[1] * x_test + x[0]
print('Фукция вида {}x + {}'.format(x[1], x[0]))
print(f_1n)
matplotlib.pyplot.plot(x_test, f_1n)
```

```python
x2 = np.array([1, 8, 15])
f2 = np.sin(x2 / 5) * np.exp(x2 / 10) + 5 * np.exp(-x2 / 2)
print(x2)
print(f2)

matrix_2n = np.array([[1, 1, 1], [1, 8, 64], [1, 15, 225]])
print(matrix_2n)

w2 = solve(matrix_2n, f2)
print(w2)
```

```python
px = np.arange(0, 20, 0.1)
py = np.sin(px / 5) * np.exp(px / 10) + 5 * np.exp(-px / 2)

px2 = np.arange(0, 20, 0.1)
py2 = w2[0] + w2[1]*px2 + w2[2]*px2*px2
matplotlib.pyplot.plot(px, py, px2, py2)
```

```python
x3 = np.array([1, 4, 10, 15])
f3 = np.sin(x2 / 5) * np.exp(x2 / 10) + 5 * np.exp(-x2 / 2)
print(x3)
print(f3)

matrix_3n = np.array([[1, 1, 1, 1], [64, 16, 4, 1], [1000, 100, 10, 1], [3375, 225, 15, 1]])
print(matrix_3n)

w3 = solve(matrix_3n, f3)
print(w3)
```

```python
px = np.arange(0, 20, 0.1)
py = np.sin(px / 5) * np.exp(px / 10) + 5 * np.exp(-px / 2)

px3 = np.arange(0, 20, 0.1)
py3 = w3[0] * (px3 ** 3) + w3[1] * (px3 ** 2) + w3[2] * px3 + w3[3]
matplotlib.pyplot.plot(px, py, px3, py3)
print('Функция имеет вид: {}x**3 + {}x**2 + {}x + {}'.format(w3[0], w3[1], w3[2], w3[3]))
```
