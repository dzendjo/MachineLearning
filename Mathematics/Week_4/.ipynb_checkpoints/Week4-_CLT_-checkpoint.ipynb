```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import math
%matplotlib inline

# Для работы используем Бета распределение
# https://en.wikipedia.org/wiki/Beta_distribution

# Возьмем параметры, при которых функция плотности распределия будет иметь интересный вид (см. график в википедии)

a=2.31
b=0.627

EX = a/(a+b) #среднее(мат ожидание, в англ. версии википедии 'mean')
DX = (a*b)/(((a+b)**2)*(a+b+1))#дисперсия (в англ. версии википедии 'variance')

print EX #m
print DX #v

#Проверим полученные значения через соответсвующие статистики
print rv.stats('mv')

# Генерируем из нашего распределения выборку объёма 1000
rv = sts.beta(a,b)
bd= rv.rvs(1000)

# Cтроим гистограмму выборки
plt.hist(bd, normed=True)

# Рисуем поверх неё теоретическую плотность распределения
x = np.linspace(0,1,1000)
pdf = rv.pdf(x)

plt.plot(x, pdf)
plt.ylabel('$f(x)$')
plt.xlabel('$x$');
```

```python
# Генерируем 1000 выборок объёма n = 5
k=1000
n=5

beta_dist_5 = rv.rvs(size=(k,n)) 

# Считаем для них выборочные средние
M_5 =[]
for i in xrange(k):
    M_5.append(0)
    for j in xrange(n):
         M_5[i] = M_5[i]+beta_dist_5[i,j]/n
            
# Строим гистограмму распределения выборочных средних
plt.hist(M_5, normed=True)
plt.ylabel('$f(x)$')
plt.xlabel('$x$')

#По ЦПТ вычислим мат ожидание и стандартное отклонение соответствующего нормального распределения
M = EX
Sg = math.sqrt(DX/n)

# Поверх гистограммы рисуем плотность соответствующего нормального распределения:

norm_rv = sts.norm(M, Sg)
x = np.linspace(0,1.5,1000)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf)
```
